```python
Diem_TB=input("Nhap diem so cua ban theo thu tu(moi dau diem cach nhau boi dau cach): ")
he_so_1, he_so_2, he_so_3 = Diem_TB.split()
a1, b1, c1=map(int,he_so_1)
a2, b2=map(int,he_so_2)
a3=int(he_so_3)
Diem_TB = ((a1+b1+c1)*1 + (a2+b2)*2 + a3*3)/ 10
print(f"Diem Trung Binh cua ban la: {Diem_TB}")
```
