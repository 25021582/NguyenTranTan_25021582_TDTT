Xep Loai Hoc Luc

```python
Hoc_Luc_Hoc_Sinh = float(input("Nhap diem trung binh cua hoc sinh:"))
if float(Hoc_Luc_Hoc_Sinh)  >= 8.0:
    print("Xep loai: Gioi")
if float(Hoc_Luc_Hoc_Sinh) >= 6.5 and float(Hoc_Luc_Hoc_Sinh) < 8.0:
    print("Xep loai: Kha")
if float(Hoc_Luc_Hoc_Sinh) >= 5.0 and float(Hoc_Luc_Hoc_Sinh) < 6.5:
    print("Xep loai: Trung Binh")
if float(Hoc_Luc_Hoc_Sinh) < 5.0:
    print("Xep loai: Yeu")        
```
